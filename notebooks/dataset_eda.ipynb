# Phase 4 Dataset EDA

Interactive exploration of the six Phase 4 datasets. Each dataset section shows:
1. Class-distribution bar chart (per-class counts across train + val + test).
2. 4×4 grid of sample images from the first four classes.
3. Image-size histogram (sampled).

Plus a side-by-side cross-region soil bar chart and an OLID I label-vocabulary table.

Run top-to-bottom with `jupyter nbconvert --to notebook --execute notebooks/dataset_eda.ipynb` to verify end-to-end.

```python
from __future__ import annotations
import sys, json
from pathlib import Path
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Make `from src...` imports work without `pip install -e .`.
PROJECT_ROOT = Path.cwd()
while not (PROJECT_ROOT / 'requirements.txt').is_file():
    PROJECT_ROOT = PROJECT_ROOT.parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from scripts._dataset_specs import DATASET_SPECS
from src.utils.data_splits import load_class_map, load_split

SPLITS = PROJECT_ROOT / 'data' / 'splits'
print('PROJECT_ROOT =', PROJECT_ROOT)
```

```python
def load_all_entries(dataset_name):
    out = {}
    for split_name in ('train', 'val', 'test'):
        p = SPLITS / dataset_name / f'{split_name}.json'
        if p.is_file():
            out[split_name] = load_split(p)
    return out

def class_counts(entries_by_split):
    total = Counter()
    for entries in entries_by_split.values():
        for e in entries:
            total[e.label] += 1
    return total

def plot_class_distribution(name, counts):
    labels = sorted(counts)
    sizes = [counts[lab] for lab in labels]
    fig, ax = plt.subplots(figsize=(max(8, len(labels) * 0.4), 4))
    ax.bar(range(len(labels)), sizes)
    ax.set_xticks(range(len(labels)))
    ax.set_xticklabels(labels, rotation=70, ha='right', fontsize=8)
    ax.set_title(f'{name} — class distribution (total={sum(sizes)})')
    ax.set_ylabel('image count')
    plt.tight_layout()
    plt.show()

def show_sample_images(name, entries, raw_root, resolver=None, n_classes=4):
    by_class = {}
    for e in entries:
        by_class.setdefault(e.label, []).append(e)
    classes = sorted(by_class)[:n_classes]
    if not classes:
        print(f'{name}: no entries to display.')
        return
    fig, axes = plt.subplots(n_classes, 4, figsize=(12, 3 * n_classes))
    if n_classes == 1:
        axes = np.array([axes])
    for row, cls in enumerate(classes):
        for col, e in enumerate(by_class[cls][:4]):
            path = resolver(e) if resolver else (raw_root / e.path)
            try:
                img = Image.open(path).convert('RGB')
                axes[row, col].imshow(img)
            except Exception as exc:
                axes[row, col].text(0.5, 0.5, f'load fail: {exc}',
                                    ha='center', va='center')
            axes[row, col].set_axis_off()
            if col == 0:
                axes[row, col].set_ylabel(cls)
    fig.suptitle(f'{name} — sample images (first {n_classes} classes)')
    plt.tight_layout()
    plt.show()

def plot_image_size_histogram(name, entries, raw_root, resolver=None, sample=200):
    sizes = []
    for e in entries[:sample]:
        path = resolver(e) if resolver else (raw_root / e.path)
        try:
            with Image.open(path) as im:
                sizes.append(im.size)
        except Exception:
            continue
    if not sizes:
        print(f'{name}: nothing to plot.')
        return
    widths = [w for w, _ in sizes]
    heights = [h for _, h in sizes]
    fig, axes = plt.subplots(1, 2, figsize=(10, 3))
    axes[0].hist(widths, bins=30); axes[0].set_title(f'{name} widths (n={len(widths)})')
    axes[1].hist(heights, bins=30); axes[1].set_title(f'{name} heights')
    plt.tight_layout()
    plt.show()
```

## plantvillage

```python
spec = next(s for s in DATASET_SPECS if s.name == 'plantvillage')
entries_by_split = load_all_entries(spec.name)
counts = class_counts(entries_by_split)
print(spec.name, 'classes=', len(counts), 'total=', sum(counts.values()))
plot_class_distribution(spec.name, counts)
```

## Cross-region soil (§14)
Side-by-side bar charts of label distributions in Phantom-fs (train+val) vs IRSID (test). Note the label spaces differ — deposit names vs texture names — which is the §14 transfer-stress signal.

```python
cr_train = load_split(SPLITS / 'soil_cross_region' / 'train.json')
cr_val = load_split(SPLITS / 'soil_cross_region' / 'val.json')
cr_test = load_split(SPLITS / 'soil_cross_region' / 'test.json')
phantom_counts = Counter(e.label for e in cr_train + cr_val)
irsid_counts = Counter(e.label for e in cr_test)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
for ax, (title, counts) in zip(axes, [('Phantom-fs (train+val)', phantom_counts),
                                       ('IRSID (test)', irsid_counts)]):
    labels = sorted(counts)
    ax.bar(range(len(labels)), [counts[l] for l in labels])
    ax.set_xticks(range(len(labels)))
    ax.set_xticklabels(labels, rotation=45, ha='right')
    ax.set_title(title)
    ax.set_ylabel('count')
plt.tight_layout()
plt.show()
```
